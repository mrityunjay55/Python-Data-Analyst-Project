##  **Exploratory Data Analysis** :

```python
sns.set(style="darkgrid")
```

```python
data.head(3)
```

### *GENDER*:_________________________________________________________

```python
data.groupby(['Gender'])['Gender'].count().reset_index(name='sum')
```

```python
ax=sns.countplot(y='Gender',data=data,palette='winter')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Count of Gender')
plt.show()
```

```python
gr=data.groupby(['Gender'])['Amount'].sum().reset_index(name='sum').sort_values(by='sum',ascending=False)
```

```python
gr
```

```python
colors=['green','red']
```

```python
plt.pie(x='sum',labels='Gender',data=gr,colors=colors,autopct='%1.1f%%')
plt.title('Amount Spent by Gender')
plt.show()
```

```python
ax=sns.barplot(x='Gender',y='sum',data=gr,palette='rocket')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')
plt.title('Amount Spent by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount');
```

### üîç ***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men...***

### *AGE & AGE GROUP*:____________________________________________________________

```python
plt.figure(figsize=(10,5))
sns.distplot(data['Age'],color='purple')
plt.title('Distribution of Age')
plt.show()
```

```python
sns.histplot(data['Age Group'],color='orange')
plt.title('Histogram of Age Group')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.show()
```

### üîç ***Here Peoples between Age of 26 To 35, do More Shopping Compare to Others***:

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x='Age Group',hue='Gender',data=data,palette='muted')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gr=data.groupby(['Age Group','Gender'])['Amount'].sum().reset_index(name='sum').sort_values(by='sum',ascending=False)
```

```python
gr
```

```python
plt.figure(figsize=(13, 6.2))
ax=sns.barplot(x='Age Group',y='sum',hue='Gender',data=gr,palette='magma')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.title('Total amount spent by Age Group and Gender')
plt.show()
```

### üîç ***From above graphs we can see that most of the buyers are of age group between 26-35 yrs female...***

### *STATE*:___________________________________________________________________

```python
data.head(2)
```

```python
gr=data.groupby(['State'])['Orders'].sum().reset_index(name='sum').sort_values(by='sum',ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
ax=sns.barplot(x='State',y='sum',data=gr,palette='inferno')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
plt.title('Top 10 States by Sum of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='State',y='Amount',data=data)
plt.title('Boxplot of Amount by State')
plt.xticks(rotation=90)
plt.show()
```

```python
gr=data.groupby(['State'])['Amount'].sum().reset_index(name='sum').sort_values(by='sum',ascending=False).head(10)
```

```python
gr
```

```python
plt.figure(figsize=(10,7))
ax = sns.barplot(y='State', x='sum', data=gr, palette='plasma')
for p in ax.patches:
    ax.annotate(format(p.get_width(), '.0f'), 
                (p.get_width(), p.get_y() + p.get_height() / 2.), 
                xytext=(5, 0), 
                textcoords='offset points', 
                ha='left', 
                va='center')

plt.title('Total Amount Spent by State')
plt.show()
```

### üîç ***From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra & Karnataka respectively...***
